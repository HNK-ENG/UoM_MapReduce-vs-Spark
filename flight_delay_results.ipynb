```python
import matplotlib.pyplot as plt
import numpy as np

iterations = np.arange(1, 6)
hadoop_times = np.array([10690, 10260, 7909, 7645, 7132])
spark_times = np.array([8033, 620, 559, 811, 423])
plt.bar(iterations - 0.2, hadoop_times, width=0.4, label='Hadoop')
plt.bar(iterations + 0.2, spark_times, width=0.4, label='Spark')
plt.xlabel('Number of Iterations')
plt.ylabel('Time Taken (ms)')
plt.title('Comparison of Hadoop and Spark Execution Times for average carrier delay')
plt.legend()
plt.savefig('comparison_plot_average carrier delay.png')
plt.show()
```
